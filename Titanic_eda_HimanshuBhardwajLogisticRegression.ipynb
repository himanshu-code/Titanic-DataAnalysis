# Logistic Regression

```python
import pandas as pd
```

```python
data=pd.read_csv('titanic_EDA.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.drop(['PassengerId','Fare','AgeGroup'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
X=data.iloc[:,1:].values
X
```

```python
y=data.iloc[:,:1].values
y
```

## Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 , random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Building The Model

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
y_pred=logmodel.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

## Model Evaluation Using Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred)
print(confusion)
```

```python
TN = confusion [0,0]
FP = confusion [0,1]
FN = confusion [1,0]
TP = confusion [1,1]
```

```python
print(confusion)
print ("TN: ", TN)
print ("FP: ", FP)
print ("FN: ", FN)
print ("TP: ", TP)
```

```python
confusion_matrix = pd.DataFrame(confusion)
confusion_matrix.columns = ['Predicted Not Survived', 'Predicted Survived']
confusion_matrix = confusion_matrix.rename(index = {0 : 'Actual Not Survived', 1 : 'Actual Survived'})
confusion_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
